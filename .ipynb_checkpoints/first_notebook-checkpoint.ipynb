```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi)
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.show()
```

```python
# List comprehension
a = [1, 2, 3, 4, 5, 6, 7]
b = [c*2 for c in a]
print(b)
```

```python
# List comprehension with if
b = [c*2 for c in a if c < 5]
print(b)
```

```python
# send tuple variable as a parameter
def add(a, b, c):
    d = a + b + c
    return d
e = (1, 2, 3)
add(e, e, e)
add(*e)
```

```python
# global variable
a = 123

def setGlobal():
    global a
    a = 456
    print("Global:", a)
    
setGlobal()
print("Global:", a)
```
