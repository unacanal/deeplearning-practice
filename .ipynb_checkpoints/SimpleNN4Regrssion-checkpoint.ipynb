```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10, 10))
```

```python
w_im = np.array([[4.0, 4.0],
                 [4.0, 4.0]]) # weights (input->middle)

w_mo = np.array([[1.0],
                 [-1.0]]) # weights (middle->output)
```

```python
b_im = np.array([3.0, -3.0])
b_mo = np.array([0.1])
```

```python
def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    print(u)
    return 1/(1+np.exp(-u)) # sigmoid
```

```python
def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return u # identity func
```

```python
for i in range(10):
    for j in range(10):
        print(i, j)
        # feed forward
        inp = np.array([X[i], Y[j]])
        print(inp)
        mid = middle_layer(inp, w_im, b_im)
        print(mid)
        out = output_layer(mid, w_mo, b_mo)
        print(out)
        
        Z[j][i] = out[0]
        print(out[0])
```

```python
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
